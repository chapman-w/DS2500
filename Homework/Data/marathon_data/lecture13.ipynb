```python
# import libraries for scraping
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
# scrape Rush's website
html = urlopen("http://rushsangs.com")
bs = BeautifulSoup(html.read(), "html.parser")
```

```python
print(bs.prettify())
```

```python
# scrape gilmore girls
html = urlopen("https://en.wikipedia.org/wiki/Gilmore_Girls")
soup = BeautifulSoup(html.read(), "html.parser")
print(soup.prettify())
```

```python
synopsis_tag = soup.find_all('h2', id='Synopsis')
p_tag = synopsis_tag[0].find_next('p')
p_tag.text
```
